```python
# Strings are Arrays - they are arrays of bytes representing unicode character
print(a[0])
print(a[-1])
```

```python
# Looping
for i in a:
  print(i)
```

## Slicing

```python
# return a range of characters by using the slice syntax.
a = "This is a string"
a[-6:-1]
```

```python
# Slice From the Start
a[:9]
```

```python
# Slice To the End
a[-6:]
```

```python
# Negative index
a[-8:]
```

## Modify Strings

```python
a
```

```python
# Upper Case
print(a.upper())
```

```python
# Lower Case
print(a.lower())
```

```python
# Remove Whitespace
a = "             Hello World               "
print(a)
print(a.strip())
print(a.rstrip())
print(a.lstrip())
```

```python
# Replace String
a = "This is a string"
print(a.replace("t","H"))
```

```python
# Split String
a = "this, is, comma, separated, string"
a.split(',')
```

## String Concatenation

```python
a = "Hello"
b = "World"
print(a+b)
```

```python
print(a+" "+b)
```

## Format - Strings

```python
# Using +
print(a+"\/"+b)
```

```python
# f - string
age = 100
print(f"This person is age {age} years")
```

### Placeholders and Modifiers

```python
# f - string
age = 50
height = 180
print(f"The age of the person is {age} and the height of that person is {height}")
```

```python
# Display the price with 2 decimals
height = 101.3986
print(f"the height of that person is {height:.2f}")
```
